```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading the Dataset

```python
df = pd.read_csv('House Price India.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df = df.rename(columns={'number of bedrooms': 'bedrooms', 'number of bathrooms': 'bathrooms', 'number of floors': 'floors', 'waterfront present':'waterfront_present', 'grade of the house': 'grade', 'lot area':'lot_area'})
```

# Univariate Analysis

```python
sns.histplot(df['Price'])
plt.title("Price Distribution")
```

```python
sns.countplot(data=df, x='grade')
plt.title('Grade Distribution')
plt.show()
```

```python
plt.pie(df.waterfront_present.value_counts(),[0,0.1],labels = ['Not Present','Present'],autopct ='%1.1f%%',shadow = True,colors = ['purple','cyan'],)
plt.title('Water Front Present Distribution')
plt.show()
```

# Bivariate Analysis

```python
sns.barplot(x=df['bedrooms'], y=df['floors'])
plt.title('Number of Floors  vs Number of Bedrooms')
plt.show()
```

```python
sns.boxplot(data=df, x='condition of the house', y='Price')
plt.title('Price by House Condition')
plt.show()
```

```python
sns.jointplot(x='lot_area', y='Price', data=df)
plt.show()
```

# Multivariate Analysis

```python
sns.pairplot(data=df, vars=['lot_area', 'bedrooms', 'bathrooms', 'Price'])
plt.show()
```

```python
df.corr()
```

```python
corr_mat = df[['Price', 'lot_area', 'Built Year', 'grade', 'Built Year','Distance from the airport']].corr()
sns.heatmap(corr_mat, annot=True, cmap='coolwarm')
```
